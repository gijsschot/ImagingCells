```python
from pylab import *
from matplotlib.colors import LogNorm
import numpy
import scipy.ndimage
import matplotlib.patches as patches
def sampling_four2real(four,cutoff=2500):
    sample = (fftshift(abs(fft(four)**2)))
    sample[sample<cutoff] = 0
    sample[sample>cutoff] = 1
    return sample

a = numpy.sin(2*numpy.pi*arange(1000)/2000/4)#random(1000)/10+1
x = 450
a[:x] = 0
a[-x:] = 0
particle = a[450:-450]
diff = fft(a)
intensities = (100*fftshift(abs(diff)**2))

critical,oversampling,undersampling = zeros(1000), zeros(1000), zeros(1000)
S = {0:critical,1:oversampling,2:undersampling}
Srate = {0:10,1:20,2:5}
for s in range(3):
    off = Srate[s]/2
    for i in arange(-off,1000+off,Srate[s]): S[s][i] = 1


sample={}
for i in range(3):
    sample[i] = sampling_four2real(S[i],2500)

fig,ax = subplots(2,4,figsize=(20,10))
fig.tight_layout()
for axs in ax:
    for a0 in axs:
        a0.set_yticks([])
        a0.set_xticks([])
        
ss = {0:"S_C",1:"S_O"}
for i in range(2):
    ax[i][0].plot(log10(intensities[450:-449]),'black');
    ax[i][1].plot(S[i*2][450:-450+1],'black')
    ax[i][2].plot(S[i*2][450:-450+1]*log10(intensities[450:-449]),'black')
    ax[i][3].plot(a,'black')
    ax[i][3].text(495, 0.19,'o', color='black', fontsize=28,backgroundcolor='white')
    ax[i][3].text(496, 0.52,'l', color='black', fontsize=28,backgroundcolor='white')
    
    ax[i][3].annotate(s='', xy=(450,.25), xytext=(550,.25), arrowprops=dict(arrowstyle='<->',linestyle='dashed'))
    ax[i][3].annotate(s='', xy=(450-100*i,.5), xytext=(550+100*i,.5), arrowprops=dict(arrowstyle='<->',linestyle='dashed'))
    
    ax[i][0].set_ylim([0,10])
    ax[i][1].set_ylim([0,2])
    ax[i][2].set_ylim([0,10])
    ax[i][3].set_ylim([0.,0.95])
    ax[i][0].text(6, 9,'Molecular Transform (MT)', color='black', fontsize=20)
    ax[i][2].text(19, 9,r'Measurement ($M$)', color='black', fontsize=20)
    ax[i][2].text(33, 8,r'$M = MT \cdot %s$' % ss[i], color='black', fontsize=20)   
    ax[i][3].text(410-60*(i-1), 0.85,'Reconstruction (R)', color='black', fontsize=20)
    ax[i][3].text(450-33*(i-1), 0.76,r'$R = DFT(M)$', color='black', fontsize=20)   
    
ax[0][1].text(15, 1.814,r'Critical Sampling ($S_C$)', color='black', fontsize=20)
ax[1][1].text(18, 1.814,r'Oversampling ($S_O$)', color='black', fontsize=20)

ax[0][3].set_xlim([450-1,1000-450])
ax[1][3].set_xlim([450-1-100*i,1000-450+1+100*i])
 
savefig('sampling.png',dpi=300)
show()
```
